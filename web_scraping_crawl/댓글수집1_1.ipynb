```python
bs=BeautifulSoup(driver.page_source,'lxml')
```

```python
li=bs.find_all('span', {'class':'u_cbox_contents'})
for i in li:
    print(i.text)
```

```python
li = bs.find_all('span', {'class': 'u_cbox_contents'})

# 파일을 저장할 경로 및 이름
file_path = r'C:\data\댓글수집.txt'

# 댓글 저장
with open(file_path, 'a', encoding='utf-8') as f:
    for i in li:
        f.write(i.text.strip() + '\n')
```
